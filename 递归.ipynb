```python
#类似于栈的先进后出模式
#1，要有递推关系
#2，要有临界
def digui(num):
    print('$' + str(num))
    if num > 0:
        #这里用的是调用本身的函数
        digui(num-1)
    else:
        print("="*20)
    print(num)
digui(3)
```

```python
#3个 7步
i = 0
def move(n,a,b,c):
    global i
    if n==1:
        i+=1
        print("移动到第",i,"次",a,'-->',c)
    else:
        move(n-1,a,c,b)
        move(1,a,b,c)
        move(n-1,b,a,c)
move(10,"A","B","C")
        
```
